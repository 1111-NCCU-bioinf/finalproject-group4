#### Performance of the eight statistical models on the testing set by variable selection method

This notebook collect the AUC score information of 8 model predictions provided by `correlation_based.ipynb`, `fold_change_based.ipynb` and `significance_ based.ipynb`, once you have execute `correlation_based.ipynb`, `fold_change_based.ipynb`, `significance_ based.ipynb` at least once, you can press `Run All` in this notebook to check the reproduction of Table 4. 

Source: https://elifesciences.org/articles/28932#table4

```python
import pandas as pd
from os import listdir
from os.path import isfile, join

path = '../results/output/AUC'
onlyfiles = [f for f in listdir(path) if isfile(join(path, f))]

AUC_table = pd.DataFrame()

display(AUC_table)

for file in onlyfiles:
    df = pd.read_csv(f'{path}/{file}')
    if file == 'correlation_based_AUC.csv':
        AUC_table['Correlation_based'] = df['AUC']
    elif file == 'fold_change_based_AUC.csv':
        AUC_table['Fold change-based'] = df['AUC']
    else:
        AUC_table['Significance-based'] = df['AUC']

AUC_table.rename(index={
    0: "Linear discriminant analysis",
    1: "Logistic regression",
    2: "Neural network",
    3: "Support vector machine",
    4: "Naive Bayes classifier",
    5: "Least Absolute Deviation regression tree",
    6: "Random forest",
    7: "Elastic net"
})
```
